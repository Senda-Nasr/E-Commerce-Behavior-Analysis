# Which gender makes more purchases online?

```python
data['Gender']= data['Gender'].replace({1:'Female', 0:'Male'})
plt.style.use('dark_background')
plt.figure(figsize=(8, 6))
sns.countplot(x='Gender', data=data, palette='viridis')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Countplot of Gender')
plt.show()
```

```python
count = data['Gender'].value_counts()
plt.style.use('dark_background')
plt.figure(figsize=(8, 8))
plt.pie(count, labels=count.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('viridis'))
plt.title('Gender Distribution')
```

# What is the distribution of purchases among the different age groups?

```python
#sns.set(style="darkgrid")
plt.style.use('dark_background')
plt.figure(figsize=(8, 6))
sns.displot(data['Age'], bins=5, kde=False, palette='viridis', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Histogram of Age')
plt.show()
```
